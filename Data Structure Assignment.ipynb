Write a code to reverse a string

````python
# Function to reverse a string
def reverse_string(s):
    return s[::-1]

# Example usage
input_string = "Hello, World!"
reversed_string = reverse_string(input_string)
print("Reversed String:", reversed_string)
````

````python
In Python, a **list** is one of the most commonly used data structures. It allows you to store an ordered collection of items, which can be of different types. Here are the key features of lists in Python:

### 1. **Ordered**  
   - Lists maintain the order of the elements as they were inserted. The order in which elements appear in a list is preserved, meaning that if you add items to a list, their order will remain unchanged unless you modify it.

### 2. **Mutable**  
   - Lists are **mutable**, meaning you can change their contents (i.e., add, remove, or modify elements) after the list has been created.

### 3. **Indexable**  
   - Elements in a list can be accessed by their **index** (i.e., their position in the list). The index starts from `0` for the first element, `1` for the second, and so on.
     ```python
     my_list = [10, 20, 30]
     print(my_list[0])  # Output: 10
     ```

### 4. **Heterogeneous**  
   - Lists can store elements of different data types (integers, strings, floats, etc.) within the same list.
     ```python
     mixed_list = [1, "hello", 3.14, True]
     ```

### 5. **Dynamic Size**  
   - Lists can grow or shrink in size during program execution. You can append elements to a list or remove them.
     ```python
     my_list = [1, 2]
     my_list.append(3)  # Adds 3 to the end of the list
     my_list.remove(2)  # Removes the first occurrence of 2
     ```

### 6. **Supports Nested Lists**  
   - Lists can contain other lists (nested lists), allowing for multi-dimensional or hierarchical data structures.
     ```python
     nested_list = [1, [2, 3], [4, 5]]
     print(nested_list[1])  # Output: [2, 3]
     ```

### 7. **Supports Slicing**  
   - Lists support **slicing**, which allows you to create a sublist by specifying a range of indices.
     ```python
     my_list = [10, 20, 30, 40, 50]
     sublist = my_list[1:4]  # Output: [20, 30, 40]
     ```

### 8. **Built-in Methods**  
   - Python lists come with a variety of useful built-in methods for operations such as adding, removing, searching, and modifying elements:
     - `append()` – Add an item to the end of the list.
     - `insert()` – Insert an item at a specific index.
     - `remove()` – Remove the first occurrence of a value.
     - `pop()` – Remove and return an item at a given index (default is the last item).
     - `extend()` – Add multiple elements to the end of the list.
     - `sort()` – Sort the list in ascending order (can be reversed with `reverse()`).
     - `reverse()` – Reverse the order of elements in the list.
     - `len()` – Return the number of elements in the list.
     ```python
     my_list = [3, 1, 2]
     my_list.sort()  # Sorts the list in-place
     print(my_list)  # Output: [1, 2, 3]
     ```

### 9. **Can Contain Duplicate Elements**  
   - Lists can contain duplicate elements. This is in contrast to other data structures like sets, which do not allow duplicates.
     ```python
     my_list = [1, 2, 2, 3]
     ```

### 10. **Supports Iteration**  
   - Lists can be iterated over using loops (like `for` loops) or comprehensions, making them very flexible for processing data.
     ```python
     my_list = [1, 2, 3]
     for item in my_list:
         print(item)
     ```

### Example of a List:
```python
fruits = ["apple", "banana", "cherry"]
print(fruits[0])  # Output: apple
fruits.append("orange")  # Adds "orange" to the end of the list
fruits[1] = "blueberry"  # Modifies the second item
print(fruits)  # Output: ['apple', 'blueberry', 'cherry', 'orange']
```

In summary, lists are versatile, ordered, and mutable collections in Python that support a wide range of operations for adding, removing, and modifying elements, as well as iteration and slicing.
````
